```python
a = pd.read_csv("Data/club_stadium_coordinates_Final_Fuzzy_Merged_2.csv")
b = pd.read_csv("Data/Transfermarkt_Cleaned.csv")
```

```python
stadium_coordinates_cleaned = a.drop('Club Name Wikidata',axis=1)
transfermarkt_cleaned_df = b.loc[:,['Name','Date','Age','Position','Nationality',
                                    'Club Left','Club Joined','Market Valuation',
                                    'Transfer','Year','Surplus/Deficit']]
```

```python
display(stadium_coordinates_cleaned)
transfermarkt_cleaned_df
```

**CREATING THE MERGED MASTER DATASETS**

```python
final_DataFrame_1 = pd.merge(transfermarkt_cleaned_df,stadium_coordinates_cleaned,
                    how='inner',
                    left_on = 'Club Left',
                    right_on='Club Name TM')

final_DataFrame_2 = pd.merge(final_DataFrame_1,stadium_coordinates_cleaned,
                    how='inner',
                    left_on = 'Club Joined',
                    right_on='Club Name TM',
                    suffixes = ('_Left','_Joined'))

final_DataFrame_2 = final_DataFrame_2.rename(columns={'Club Name TM_Joined':'Club_TM_Joined', 
                                                      'Club Name TM_Left':'Club_TM_Left',
                                                      'venueLabel_Left':'Stadium_Left',
                                                      'venueLabel_Joined':'Stadium_Joined'})

final_DataFrame_2 = final_DataFrame_2.loc[:,['Name','Date','Age','Position','Nationality',
                                             'Club Left','Stadium_Left','Latitude_Left','Latitude_Joined',
                                             'Club Joined','Stadium_Joined','Latitude_Joined','Longitude_Joined',
                                             'Market Valuation','Transfer','Year','Surplus/Deficit']]

final_DataFrame_2
```

**CREATING THE DATA SET FOR THE NODES**

```python
Transfer_Value = final_DataFrame_2.groupby('Club Joined').agg({'Transfer':'sum'}).reset_index()

Transfer_Surplus = final_DataFrame_2.groupby('Club Joined').agg({'Surplus/Deficit':'sum'})

agg_surplus_val = pd.merge(Transfer_Surplus,Transfer_Value,
                    how='outer',
                    on = 'Club Joined')

merged_df_2 = pd.merge(stadium_coordinates_cleaned,agg_surplus_val,
                    how='outer',
                    left_on = 'Club Name TM',
                    right_on = 'Club Joined')

merged_df_2 = merged_df_2.fillna(0)

merged_df_2['Transfer'] = merged_df_2['Transfer']

merged_df_2 
```
